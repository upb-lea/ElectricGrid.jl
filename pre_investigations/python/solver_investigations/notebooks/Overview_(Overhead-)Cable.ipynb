# Kabel

```julia
using ControlSystems, Plots,DisplayAs
```

```julia
R_K = 0.772
L_K = 0.083/(2*pi*50)
C_K = 0.4 * 10^-6
G_K = 0
l1= 1
l2= 10
l3= 100
```

```julia
x = []
#1km
push!(x,2)
push!(x,R_K*C_K*l1^2+L_K*G_K*l1^2)
push!(x,L_K*C_K*l1^2)

#10km
push!(x,2)
push!(x,R_K*C_K*l2^2+L_K*G_K*l2^2)
push!(x,L_K*C_K*l2^2)

#100km
push!(x,2)
push!(x,R_K*C_K*l3^2+L_K*G_K*l3^2)
push!(x,L_K*C_K*l3^2)
```

```julia
setPlotScale("dB")
P = tf([2.0],[x[9],x[8],x[7]])
    fig = bodeplot(P,hz=true,label= "100km")
P = tf([2.0],[x[6],x[5],x[4]])
    fig = bodeplot!(P,hz=true,label= "10km",colour="red")
P = tf([2.0],[x[3],x[2],x[1]])
    fig = bodeplot!(P,hz=true,label= "1km")
```

```julia
P = tf([2.0],[x[3],x[2],x[1]])
    fig = bodeplot(P,hz=true,label= "1km",colour="green")
```

```julia
P = tf([2.0],[x[6],x[5],x[4]])
    fig = bodeplot(P,hz=true,label= "10km",xscale = :log10,colour="red")
```

```julia
P = tf([2.0],[x[9],x[8],x[7]])
    fig = bodeplot(P,hz=true,label= "100km")
```
